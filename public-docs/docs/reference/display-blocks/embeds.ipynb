::: datapane.Media

### Simple image embed

In the following example, Datapane will display images in your app for viewers and allow users to download them.

```python
dp.Media(file="./image.png", name="Image1", caption="Datapane in action!")
```

::: datapane.Embed

### Simple YouTube embed

```python
dp.Embed(url='https://www.youtube.com/watch?v=_KS_yZBI71s&t')
```

You don't need to use this block for simple embeds, e.g. GIFs. For those, just use Markdown syntax i.e. `![](https://my-example-gif.gif)`

::: datapane.Attachment

### Simple attachment

```python
data_path = "./netflix_stock_data.csv"

dp.Attachment(file=data_path)
```

### Python object attachment

```python
vehicle_dict = {"brand": "Ford", "model": "Mustang", "year": 1964}

dp.Attachment(vehicle_dict, name="vehicle_dict")
```

### Multiple attachments

```python
data_path = "./netflix_stock_data.csv"

vehicle_dict = {"brand": "Ford", "model": "Mustang", "year": 1964}

dp.Group(
    dp.Attachment(file=data_path), dp.Attachment(vehicle_dict, name="vehicle_dict")
)
```
