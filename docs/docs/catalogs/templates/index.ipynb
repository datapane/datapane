# Templates catalog

Welcome to the Datapane templates catalog. Here you will find expertly crafted Datapane templates that can be used as a starting point for your next project.

To get up and running with a template, you can:

- :octicons-mark-github-16: Visit the template repository to browse the source code.

- :octicons-terminal-16: Use the `datapane template` command to retrieve and start up a template.

- :octicons-repo-clone-16: Clone the repository directly.

## Custom templates

Custom Datapane templates can be created by following the instructions on [dp-template](https://github.com/datapane/dp-template) repository.

You can create instances of this template by using the `datapane template` command followed by the template repository URL, e.g.:

```bash
datapane template https://github.com/datapane/dp-template-classifier-dashboard.git
```

By default, the `datapane template` command automatically executes and presents the template app. This can be disabled with the `--no-execute` option, e.g.:

```bash
datapane template https://github.com/datapane/dp-template-classifier-dashboard.git --no-execute
```
